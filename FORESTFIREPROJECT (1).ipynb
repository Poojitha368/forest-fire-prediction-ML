```python
d=x.drop('month',axis=1)
d1=d.drop('day',axis=1)
d1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(d1,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#linear regression
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression()

model.fit(x_train, y_train)


y_pred = model.predict(x_test)


mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)


print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2):", r2)
```

```python
#Ridge regression
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=1.0)


ridge_model.fit(x_train, y_train)


y_ridge_pred = ridge_model.predict(x_test)

ridge_mse = mean_squared_error(y_test, y_ridge_pred)
ridge_rmse = np.sqrt(ridge_mse)
ridge_r2 = r2_score(y_test, y_ridge_pred)



print("Ridge Mean Squared Error (MSE):", ridge_mse)
print("Ridge Root Mean Squared Error (RMSE):", ridge_rmse)
print("Ridge R-squared (R2):", ridge_r2)
```

```python
#lasso regression
from sklearn.linear_model import Lasso


lasso_model = Lasso(alpha=1.0)


lasso_model.fit(x_train, y_train)


y_lasso_pred = lasso_model.predict(x_test)


lasso_mse = mean_squared_error(y_test, y_lasso_pred)
lasso_rmse = np.sqrt(lasso_mse)
lasso_r2 = r2_score(y_test, y_lasso_pred)


print("Lasso Mean Squared Error (MSE):", lasso_mse)
print("Lasso Root Mean Squared Error (RMSE):", lasso_rmse)
print("Lasso R-squared (R2):", lasso_r2)
```

```python
# support vector machine
from sklearn.svm import SVR


svm_model = SVR(kernel='linear', C=1.0, epsilon=0.2)


svm_model.fit(x_train, y_train)


y_svm_pred = svm_model.predict(x_test)


svm_mse = mean_squared_error(y_test, y_svm_pred)
svm_rmse = np.sqrt(svm_mse)
svm_r2 = r2_score(y_test, y_svm_pred)


print("SVM Mean Squared Error (MSE):", svm_mse)
print("SVM Root Mean Squared Error (RMSE):", svm_rmse)
print("SVM R-squared (R2):", svm_r2)
```

```python
# knn regression
from sklearn.neighbors import KNeighborsRegressor


knn_model = KNeighborsRegressor(n_neighbors=5)


knn_model.fit(x_train, y_train)


y_knn_pred = knn_model.predict(x_test)


knn_mse = mean_squared_error(y_test, y_knn_pred)
knn_rmse = np.sqrt(knn_mse)
knn_r2 = r2_score(y_test, y_knn_pred)


print("KNN Mean Squared Error (MSE):", knn_mse)
print("KNN Root Mean Squared Error (RMSE):", knn_rmse)
print("KNN R-squared (R2):", knn_r2)
```

```python
# bootstrapping for linear regression
from sklearn.utils import resample
import numpy as np
from sklearn.linear_model import LinearRegression

num_iterations = 100

# Initialize a list to store MSE values
mse_values = []

# Perform bootstrapping
for _ in range(num_iterations):
    # Resample the training data with replacement
    X_resampled, y_resampled = resample(x_train, y_train)

    # Create and train a model (e.g., Linear Regression)
    model = LinearRegression()
    model.fit(X_resampled, y_resampled)

    # Make predictions on the test data
    y_pred = model.predict(x_test)

    # Calculate the mean squared error
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)

# Calculate the mean and standard deviation of MSE values
mean_mse = sum(mse_values) / len(mse_values)
std_mse = (sum((mse - mean_mse) ** 2 for mse in mse_values) / (len(mse_values) - 1)) ** 0.5
linear=mean_mse

# Print results
print(f"Mean MSE: {mean_mse}")
print(f"Standard Deviation of MSE: {std_mse}")
```

```python
c=range(100)
plt.title('Bootstrapped MSE Distribution for Linear Regression')
plt.xlabel('MSE Values')
plt.ylabel('Frequency')
plt.plot(c,mse_values)
```

```python
# perform ridge regression
from sklearn.linear_model import Ridge
from sklearn.utils import resample

# Set the number of bootstrap iterations
num_iterations = 100

# Initialize a list to store MSE values
mse_values = []

# Perform bootstrapping
for _ in range(num_iterations):
    # Resample the training data with replacement
    X_resampled, y_resampled = resample(x_train, y_train)

    # Create and train a Ridge Regression model
    model = Ridge(alpha=1.0)  # Adjust alpha as needed
    model.fit(X_resampled, y_resampled)

    # Make predictions on the test data
    y_pred = model.predict(x_test)

    # Calculate the mean squared error
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)

# Calculate the mean and standard deviation of MSE values
mean_mse = sum(mse_values) / len(mse_values)
std_mse = (sum((mse - mean_mse) ** 2 for mse in mse_values) / (len(mse_values) - 1)) ** 0.5

ridge=mean_mse
# Print results
print(f"Mean MSE for Ridge Regression: {mean_mse}")
print(f"Standard Deviation of MSE: {std_mse}")
```

```python
c=range(100)
plt.title('Bootstrapped MSE Distribution for Ridge Regression')
plt.xlabel('MSE Values')
plt.ylabel('Frequency')
plt.plot(c,mse_values)
```

```python
# perform bootstrapping for lasso regression
from sklearn.linear_model import Lasso
from sklearn.utils import resample

# Set the number of bootstrap iterations
num_iterations = 100

# Initialize a list to store MSE values
mse_values = []

# Perform bootstrapping
for _ in range(num_iterations):
    # Resample the training data with replacement
    X_resampled, y_resampled = resample(x_train, y_train)

    # Create and train a Lasso Regression model
    model = Lasso(alpha=1.0)  # Adjust alpha as needed
    model.fit(X_resampled, y_resampled)

    # Make predictions on the test data
    y_pred = model.predict(x_test)

    # Calculate the mean squared error
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)

# Calculate the mean and standard deviation of MSE values
mean_mse = sum(mse_values) / len(mse_values)
std_mse = (sum((mse - mean_mse) ** 2 for mse in mse_values) / (len(mse_values) - 1)) ** 0.5

lasso=mean_mse
# Print results
print(f"Mean MSE for Lasso Regression: {mean_mse}")
print(f"Standard Deviation of MSE: {std_mse}")
```

```python
c=range(100)
plt.title('Bootstrapped MSE Distribution for Lasso Regression')
plt.xlabel('MSE Values')
plt.ylabel('Frequency')
plt.plot(c,mse_values)
```

```python
#perform bootstrapping for knn
from sklearn.neighbors import KNeighborsRegressor
from sklearn.utils import resample

# Set the number of bootstrap iterations
num_iterations = 100

# Initialize a list to store MSE values
mse_values = []

# Perform bootstrapping
for _ in range(num_iterations):
    # Resample the training data with replacement
    X_resampled, y_resampled = resample(x_train, y_train)

    # Create and train a KNN Regression model
    model = KNeighborsRegressor(n_neighbors=5)  # Customize the number of neighbors as needed
    model.fit(X_resampled, y_resampled)

    # Make predictions on the test data
    y_pred = model.predict(x_test)

    # Calculate the mean squared error
    mse = mean_squared_error(y_test, y_pred)
    r2=r2_score(y_test, y_pred)
    mse_values.append(mse)


mean_mse = sum(mse_values) / len(mse_values)
std_mse = (sum((mse - mean_mse) ** 2 for mse in mse_values) / (len(mse_values) - 1)) ** 0.5

knn=mean_mse
print(f"Mean MSE for KNN Regression: {mean_mse}")
print(f"Mean r2 for KNN Regression: {r2}")
print(f"Standard Deviation of MSE: {std_mse}")
```

```python
c=range(100)
plt.title('Bootstrapped MSE Distribution for  KNN Regression')
plt.xlabel('MSE Values')
plt.ylabel('Frequency')
plt.plot(c,mse_values)
```

```python
# perform bootstrapping for svm
from sklearn.svm import SVR
from sklearn.utils import resample


num_iterations = 100


mse_values = []

for _ in range(num_iterations):
    X_resampled, y_resampled = resample(x_train, y_train)

    model = SVR(kernel='linear', C=1.0, epsilon=0.2)
    model.fit(X_resampled, y_resampled)


    y_pred = model.predict(x_test)


    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)


mean_mse = sum(mse_values) / len(mse_values)
std_mse = (sum((mse - mean_mse) ** 2 for mse in mse_values) / (len(mse_values) - 1)) ** 0.5

svm=mean_mse
# Print results
print(f"Mean MSE for SVM Regression: {mean_mse}")
print(f"Standard Deviation of MSE: {std_mse}")
```

```python
c=range(100)
plt.title('Bootstrapped MSE Distribution for support vector machine')
plt.xlabel('MSE Values')
plt.ylabel('Frequency')
plt.plot(c,mse_values)
```
